## 2.Reading excel files

```python
import re

phoneNum = re.compile(r'0\d\d\d\d\d\d\d\d\d*')

def convert_studentNumber(cell):
    """if you wanna do conversion of your cell content from excel"""
    isPhone = phoneNum.search(cell)
    if isPhone is not None:
        return 'invalid input'
    return cell
```

```python
df = pd.read_excel("data/Students_Details.xlsx","Unverified Students",skiprows=2, converters = {
    'Student /Staff Number': convert_studentNumber
})
```

```python
display(df[21:67])
```

#### writing two dataframes to same excel file as sheet 1 and sheet 2

we use a class called pd.ExcelWriter

```python
#lets first create two data frames
df_stocks = pd.DataFrame({
    'tickers':['GooGl','WMT','MSFT'],
    'price':[845,65,64],
    'pe':[30.37, 14.26, 30.97],
    'eps':[27.82,4.61,2.12]
})
df_weather = pd.DataFrame({
    'day':['1-1-2016','1-2-2016','1-3-2016','1-4-2016','1-5-2016','1-6-2016'],
    'temperature':[32,35,28,29,32,31],
    'windspeed':[6,7,2,7,4,2],
})
```

```python
with pd.ExcelWriter('data/stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name='stocks',index=False)
    df_weather.to_excel(writer, sheet_name='weather',index=False)
```
