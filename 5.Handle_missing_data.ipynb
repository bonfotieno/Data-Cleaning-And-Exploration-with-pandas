# Handling missing data

```python
import pandas as pd
df = pd.read_csv("data\weather_data.csv", parse_dates=["day"]) #parse_dates changes the day column from str to date type
df.set_index('day', inplace=True) #inplace = True so that it can modify the df
df
```

### using _fillna_

```python
new_df = df.fillna(0)
new_df
```

#### specifying different values for different columns

```python
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

#### filling with some reasonable data

```python
new_df = df.fillna(method='ffill', limit=2) #foward fill, carry foward the previous value | limit to specify number of cells to copy to.
new_df
```

```python
new_df = df.fillna(method='bfill', axis='rows') #backward fill, axis specify either columns or rows
new_df
```

### _using interpolate_
comes up with a better guess using linear interpolation

```python
new_df = df.interpolate() #but here we use linear interpolation
new_df
```

##### time interpolation

```python
new_df = df.interpolate(method='time')
new_df
```

### Droping all rows with na values

```python
new_df = df.dropna()
new_df
```

#### Drop if the entire row is na values

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=2) # keep row if has atleast two of the values
new_df
```

### Inserting  the missing dates

```python
dt = pd.date_range("2017-01-01","2017-01-11")
index = pd.DatetimeIndex(dt)
df = df.reindex(index)
df
```

### using replace function

```python
import numpy as np
new_df = df.replace('T',np.NaN) # instead of a single value you can supply a list ['T',-888888]
new_df
```

##### replacing based on specific column
<pre><code>
    new_df = df.replace({
        'temperature':'Y', 
        'windspeed':'T',
        'event':'no event'
    },np.NaN) 
    new_df
</code></pre>
##### replacing based on specific values
<pre><code>
    new_df = df.replace({
        ['Y','T']:np.NaN,
        'no event':'sunny'
    })
    new_df
</code></pre>

##### replacing using regex
<pre><code>
    new_df = df.replace({
        'temperature':'[A-Za-z]',
        'windspeed':'[A-Za-z]'
    },'',regex=True)
</code></pre>

##### replacing a list of  values with a list of values

```python
df = pd.DataFrame(
    {
        'score':['exceptional','average','good','poor','average','exceptional'],
        'student':['rob','maya','parthiv','tom','julian','erica']
    }
)
df
```

```python
new_df = df.replace(['poor','average','good','exceptional'],[1,2,3,4])
new_df
```
